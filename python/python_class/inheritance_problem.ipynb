### 문제 1
동물을 나타내는 `Animal` 클래스와 이를 상속받는 `Dog`, `Cat` 클래스를 작성하세요.
1. `Animal` 클래스는 생성자(`__init__`)에서 이름(`name`)을 저장하고, `move(self)` 메서드에서 "[이름]이 이동합니다."를 출력합니다.
2. `Dog` 클래스는 `Animal`을 상속받으며, 고유 메서드 `bark(self)`에서 "[이름]이 멍멍 짖습니다."를 출력합니다.
3. `Cat` 클래스는 `Animal`을 상속받으며, 고유 메서드 `mew(self)`에서 "[이름]이 냥냥 웁니다."를 출력합니다.
4. `Dog`, `Cat` 인스턴스를 생성하고 두 메서드를 모두 호출해보세요.

```python
# 여기에 코드를 작성하세요.

class Animal:
    def __init__(self, name):
        self.name = name
    
    def move(self):
        print(f'{self.name}이/가 이동합니다.')

    def eat(self):
        print('먹는다!')

class Dog(Animal):
    def bark(self):
        print(f'{self.name} 짖는다!')
    
    def eat(self):
        super().eat()
        print("개가 먹는다!")

class Cat(Animal):
    def mew(self):
        print(f'{self.name}이 웁니다')

a = Animal('초코')
a.move()
dog = Dog('흰둥이')
dog.move()
dog.bark()
cat = Cat('고냥이')
cat.move()
cat.mew()
dog.eat()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        # 인스턴스 변수를 활용한 정보 출력
        print(f"이름은 {self.name}이고 나이는 {self.age}세입니다.")

    def celebrate_birthday(self):
        # 인스턴스 변수 age의 값을 1 증가시켜 상태 변경
        self.age += 1

person = Person("박지성", 40)
# 인스턴스 메서드 호출
person.introduce()
person.celebrate_birthday()
print(person.age)
```

```python
class Address:
    def __init__(self, city, street):
        # 주소 정보를 저장하는 인스턴스 변수
        self.city = city
        self.street = street

    def get_full_address(self):
        # 상세 주소를 문자열로 반환하는 메서드
        return f"{self.city} {self.street}"

class Person:
    def __init__(self, name, city, street):
        self.name = name
        # Address 클래스의 인스턴스를 생성하여 포함 관계 형성
        self.address = Address(city, street)

    def print_info(self):
        # 포함된 Address 객체의 메서드를 호출하여 데이터 처리
        full_addr = self.address.get_full_address()
        print(f"이름: {self.name}, 주소: {full_addr}")

# Person 인스턴스 생성 시 내부에서 Address 인스턴스도 함께 생성됨
p = Person("홍길동", "서울시", "강남대로")
p.print_info()
```
