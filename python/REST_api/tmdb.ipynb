### 문제 1
현재 상영 중인 영화(Now Playing) 중 평점(vote_average)이 가장 높은 영화를 찾으시오.

```python
import requests
from pprint import pprint

URL = 'https://api.themoviedb.org/3/movie/now_playing'
API_KEY = os.getenv('TMDB_API_KEY')

params = {
    'language' : 'ko-kr',
}

headers = {
    "Authorization" : f"Bearer {API_KEY}"
}

try:
    response = requests.get(URL, headers=headers, params=params)

    response.raise_for_status()
    
    data = response.json()
    data = data['results']
    max_title = ""
    max_vote_average = 0
    for movie in data:
        title = movie['title']
        vote_average = movie['vote_average']
        if max_vote_average < vote_average:
            max_vote_average = vote_average
            max_title = title
    
    pprint(max_title)
    
except Exception as e:
    print(e)
```

### 문제 3
현재 상영 중인 영화(now_playing) 데이터에서 다음 정보만 담긴 리스트를 만드세요.

- title (영화 제목)  
- vote_average (평점)  
- release_date (개봉일)

```python
URL = "https://api.themoviedb.org/3/movie/now_playing"
API_KEY = os.getenv('TMDB_API_KEY')

params = {
    'language' : 'ko-kr',
}

headers = {
    "Authorization" : f"Bearer {API_KEY}"
}

try:
    response = requests.get(URL, headers=headers, params=params)

    response.raise_for_status()
    
    data = response.json()
    data = data['results']
    movie_lst = []
    for movie in data :
        title = movie['title']
        vote_average = movie['vote_average']
        release_date = movie['release_date']
        info = {
            "title" : title,
            'vote_average' : vote_average, 
            'release_date' : release_date
            }
        movie_lst.append(info)
    pprint(movie_lst)
    
except Exception as e:
    print(e)
```

### 문제 4
현재 상영 중인 영화(now_playing) 데이터를 vote_average(평점)순으로 정렬하여 영화 제목이 담긴 리스트를 만드세요.
단, 리스트의 길이는 100 이상이어야 합니다.


```python
import requests
from pprint import pprint

URL = "https://api.themoviedb.org/3/movie/now_playing"
API_KEY = os.getenv('TMDB_API_KEY')

params = {
        'language' : 'ko-kr', 
    }

headers = {
        "Authorization" : f"Bearer {API_KEY}"
    }

try:
    all_movies = []
    for page_num in range(1, 6):
        params['page'] = page_num
        response = requests.get(URL, headers=headers, params=params)
        response.raise_for_status()
            
        data = response.json()
        data = data['results']
        all_movies.extend(data)
    sorted_movies = sorted(all_movies, key=lambda el : el ['vote_average'], reverse=True)
    final_titles = [movie['title'] for movie in sorted_movies]
        # 내가한거
        # for movie in sorted_movies:
        #     title = movie['title']
        #     movie_titles.append(title)

    pprint(final_titles)
    print(f"총 {len(final_titles)}개의 영화 제목을 가져왔습니다.")
    # lst = [movie['title'] for movie in data1] 리스트안에 포문을넣어서 한줄로 끝낼수도 있음
except Exception as e:
    print(e)
```
